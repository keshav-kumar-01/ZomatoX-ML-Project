```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Step 2: Load Dataset
df = pd.read_csv('D:\ZomatoX-ML\Data\enhanced_zomato_dataset_clean.csv')
df.head()
```

```python
df.tail()
```

```python
# Shape and column overview
print(f"Shape: {df.shape}")
df.info()
df.describe()

# Check for nulls (should be clean, but we confirm)
df.isnull().sum()
```

```python
# Top 10 cities by number of restaurants
top_cities = df['City'].value_counts().nlargest(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_cities.index, y=top_cities.values)
plt.title('Top 10 Cities by Restaurant Count')
plt.xticks(rotation=45)
plt.show()

# Rating distribution
sns.histplot(df['Average_Rating'], bins=20, kde=True)
plt.title("Distribution of Average Ratings")
plt.show()

# Cuisine popularity
top_cuisines = df['Cuisine'].value_counts().nlargest(10)
sns.barplot(y=top_cuisines.index, x=top_cuisines.values)
plt.title('Top 10 Cuisines')
plt.xlabel('Number of Restaurants')
plt.show()

# Price distribution
sns.boxplot(x='Is_Expensive', y='Prices', data=df)
plt.title("Price vs Expensive Flag")
```

```python
# Value for money by city
city_price_vote = df.groupby("City")[['Price_per_Vote', 'Avg_Rating_City']].mean().sort_values(by='Price_per_Vote')
city_price_vote.plot(kind='bar', figsize=(14,5), title='Price per Vote & Avg Rating by City')

# Bestseller % per city
bestseller_ratio = df.groupby('City')['Is_Bestseller'].mean().sort_values(ascending=False).head(10)
bestseller_ratio.plot(kind='barh', title='Top Cities by % of Bestseller Items')
```
